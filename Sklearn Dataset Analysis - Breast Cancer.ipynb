```python
# Getting data into dataframe
import pandas as pd
data = pd.DataFrame(cancer_ds.data, columns=cancer_ds.feature_names)
# The column=feature_names, to name the columns as features
data
```

```python
# Appending a column called class (which is the target {0,1})
data['class'] = cancer_ds.target
```

```python
#Finding how many malignant and benign cancer samples
print(data['class'].value_counts())
```
