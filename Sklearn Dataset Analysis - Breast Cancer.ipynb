```python
# Train-Test Split
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('class', axis=1)
Y = data['class']
```

```python
# Splitting the training and testing data by first shuffling it and giving 10% of data as the test dataset.
# Can also do cross-fold validation, by dividing test and train set into various folds.

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.1, random_state=10)
print(X_train.shape, Y_train.shape)
print(X_test.shape, Y_test.shape)
```

```python
mean_radius_max =  X_train['mean radius'].max()
mean_radius_min =  X_train['mean radius'].min()
mean_radius_half =  (X_train['mean radius'].max() +  X_train['mean radius'].min()) / 2.0
print(f'Mean Radius Max: {mean_radius_max}')
print(f'Mean Radius Min: {mean_radius_min}')
print(f'Mean Radius Half: {mean_radius_half}')
```

```python
# Pandas.cut() -> Used cut when you need to segment and sort data values into bins/categories. 
# This function is useful for going from a continuous variable to a categorical variable. 

X_train_binarised = X_train.apply(pd.cut, bins=2, labels=[0,1])
```

**NOTE:** Here the bins=2, meaning we are cutting the X_train into 2 categories. So example, if we looks at "Mean Radius", where the half is 17.5455. So the mean radius < 17.5455, would be labelled as 0 and the other half > 17.5455 would be labelled as 1. (0,1 comes from the label parameter.)

```python
# ML Model
from random import randint

# NOTE: Here the model tells if the cancer is malignant or benign depending on the number of features contributing towards the prediction.
# Remember num of features as we have converted everything to be categorical.
# VERY INACURATE way of prediction :(
# b is the thresh-hold of the number of features to account for if the cancer is malignant or benign.
# b would be in the range 0 to 30 (max no. of features.)
b = 10 

i = randint(0, X_train.shape[0])
print(i)
print(X_train_binarised[i])
if (np.sum(X_train_binarised[i,:])>=b):
    print("ML Model says its Malignant.")
else:
    print("ML Model says its Benign.")
    
if (Y_train[i] == 1):
    print("Actual Output is Malignant.")
else:
    print("Actual Output is Benign.")
```

## Checking accuracy for different values of b.

```python
print()
```

```python
def get_accuracy(b):
    count = 0
    correct_count = 0
    for i in range(len(X_train_binarised)):
        if (np.sum(X_train_binarised[i,:])>=b):
            pred_output=1
        else:
            pred_output=0

        if (Y_train.iloc[i] == 1):
            actual_output=1
        else:
            actual_output=0
            
        if (pred_output==actual_output):
            correct_count+=1
        count+=1
    return (correct_count*100/count)
```

```python
for b in range(32):
    acc = get_accuracy(b)
    print(f'Value of b: {b} \t Accuracy: {acc}')
```
